```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Verificar si tenemos GPU disponible (opcional pero buena práctica)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Usando el dispositivo: {device}")
```

```python
# Crear datos sintéticos
X = torch.linspace(0, 10, 100).reshape(-1, 1)  # 100 puntos de 0 a 10
noise = torch.randn(X.shape) * 2               # Un poco de ruido aleatorio
y = 3 * X + 2 + noise                          # La "verdad" que la red debe aprender

# Graficar para ver qué tenemos
plt.scatter(X, y)
plt.title("Datos de prueba para nuestra Red")
plt.show()
```

```python
class MiPrimeraRed(nn.Module):
    def __init__(self):
        super(MiPrimeraRed, self).__init__()
        # Definimos una capa lineal: 1 entrada (X) y 1 salida (y)
        self.capa_lineal = nn.Linear(1, 1)

    def forward(self, x):
        # Aquí definimos cómo fluyen los datos a través de la red
        return self.capa_lineal(x)

# Instanciar el modelo
modelo = MiPrimeraRed().to(device)
```

```python
criterio = nn.MSELoss()
optimizador = optim.SGD(modelo.parameters(), lr=0.01) # lr = learning rate
```

```python
epochs = 100 # Cuántas veces verá la red los datos

for epoch in range(epochs):
    # 1. Forward pass
    predicciones = modelo(X)
    
    # 2. Calcular el error
    loss = criterio(predicciones, y)
    
    # 3. Limpiar gradientes
    optimizador.zero_grad()
    
    # 4. Backpropagation
    loss.backward()
    
    # 5. Ajustar pesos
    optimizador.step()
    
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Error: {loss.item():.4f}')
```

```python
# Desactivamos el cálculo de gradientes para graficar (ahorra memoria)
with torch.no_grad():
    predicciones_finales = modelo(X).cpu()

plt.scatter(X, y, label='Datos reales')
plt.plot(X, predicciones_finales, color='red', label='Predicción de la Red')
plt.legend()
plt.show()
```
